## Matplotlib Basics

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,5,11)
y=x**2
```

plt.plot(x,y)
plt.title("First Graph")
plt.xlabel("x axis")
plt.ylabel("y axis")

## SubPlots

```python
# (rows ,cols,position of block)
plt.subplot(2,2,1)
# to create plot we just write below subplot what we want to plot 
plt.plot(x,y)

plt.subplot(2,2,2)
plt.plot(y,x)
plt.subplot(2,2,3)
plt.plot(y,y)
plt.subplot(2,2,4)
plt.plot(x,y**2)
```
