```python
import numpy as np 
import pandas as pd
```

## Series 


```python
my_list=[10,20,30];
my_arr=np.array(my_list);
pd.Series(my_arr)
# Here left side 0 1 2 acting as label
```

```python
labels=['a','b','c'];
pd.Series(my_arr,index=labels)
# above is the way of using custom labels
```

```python
# With dictionary key automatically act as a label
d={'a':10,'b':20,'c':30}
pd.Series(d)
```
