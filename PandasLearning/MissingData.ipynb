## Finding Missing Data

```python
import numpy as np 
import pandas as pd
```

```python
data = {
    'A': [1, 2, None, 4, 5],
    'B': [1, 2, 3, 4, 5],
    'C': [1, 2, 3, np.nan, np.nan],
    'D': [1, np.nan, np.nan, np.nan, 5]
}
df = pd.DataFrame(data)
```

```python
df
```

```python
# this will check for missing value
df.isna()
```

```python
df.isna().sum()
```

```python
df.isna().any()
```

## Remove Missing Value

```python
df=df.dropna()
#df=df.dropna(axis=1) # for column 
# this will delete all rows which has missing value
```

```python
df
```

```python
df.dropna(thresh=3)
# thresh this told atleast this non na number 
```

```python
df
```

## Fill the missing data

```python
df.fillna(0,inplace=True)
```

```python
df
```

```python
df
```

```python
#custom values column wise 
values = {'A':0,'B':100,"C":300,'D':400}
df=df.fillna(value=values)
df
```
